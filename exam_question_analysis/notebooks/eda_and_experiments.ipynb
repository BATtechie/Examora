# Examora — Milestone 1: EDA & Experiments

Exploratory Data Analysis of `exam_question_dataset_5000.csv`

Steps covered:
1. Data Loading & Overview
2. Class Distribution
3. Numerical Feature Distributions
4. Categorical Feature Analysis
5. Text Feature Exploration
6. Correlation Analysis
7. Baseline Experiments

````python
import sys, os
sys.path.insert(0, os.path.join(os.getcwd(), '..'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid', palette='muted')
%matplotlib inline

from src.preprocessing import load_data, clean_text

DATA_PATH = '../data/exam_question_dataset_5000.csv'
df = load_data(DATA_PATH)
df.head()
````

````python
# ── 1. Shape & Types
print('Shape:', df.shape)
print()
df.dtypes
````

````python
# ── 2. Missing Values
df.isnull().sum()
````

````python
# ── 3. Class Distribution
fig, ax = plt.subplots(figsize=(6, 4))
df['difficulty_label'].value_counts().plot(kind='bar', ax=ax, color=['#22c55e','#f59e0b','#ef4444'])
ax.set_title('Difficulty Label Distribution', fontsize=13)
ax.set_xlabel('Difficulty')
ax.set_ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

````python
# ── 4. Numerical Feature Distributions
num_cols = ['avg_score', 'std_dev', 'discrimination_index']
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, col in zip(axes, num_cols):
    sns.histplot(data=df, x=col, hue='difficulty_label', kde=True, ax=ax,
                 palette={'Easy':'#22c55e','Medium':'#f59e0b','Hard':'#ef4444'})
    ax.set_title(col)
plt.suptitle('Numerical Feature Distributions by Difficulty', y=1.02, fontsize=13)
plt.tight_layout()
plt.show()
````

````python
# ── 5. Categorical Features
cat_cols = ['subject', 'question_type', 'cognitive_level']
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
for ax, col in zip(axes, cat_cols):
    ct = pd.crosstab(df[col], df['difficulty_label'], normalize='index') * 100
    ct.plot(kind='bar', stacked=True, ax=ax,
            color=['#22c55e','#ef4444','#f59e0b'])
    ax.set_title(f'{col} vs Difficulty (%)')
    ax.set_ylabel('Percentage')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right')
plt.tight_layout()
plt.show()
````

````python
# ── 6. Correlation Heatmap (numerical)
from sklearn.preprocessing import LabelEncoder
df_corr = df[num_cols + ['difficulty_label']].copy()
df_corr['difficulty_encoded'] = LabelEncoder().fit_transform(df_corr['difficulty_label'])
df_corr.drop(columns='difficulty_label', inplace=True)

fig, ax = plt.subplots(figsize=(5, 4))
sns.heatmap(df_corr.corr(), annot=True, fmt='.2f', cmap='coolwarm', ax=ax, linewidths=0.5)
ax.set_title('Correlation Heatmap')
plt.tight_layout()
plt.show()
````

````python
# ── 7. Text Length Distribution
df['text_length'] = df['question_text'].str.split().str.len()
print(df['text_length'].describe())
df['text_length'].hist(bins=25, color='#6366f1', edgecolor='white')
plt.title('Question Text Length (words)')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show()
````

````python
# ── 8. Sample Cleaned Text
df['cleaned_text_sample'] = df['question_text'].apply(clean_text)
df[['question_text', 'cleaned_text_sample']].head(5)
````
